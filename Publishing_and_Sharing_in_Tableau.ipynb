What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/
Cloud in terms of accessibility and data security?

    ->Accessibility
    -Tableau Public dashboards are fully public anyone online can view, interact with, and download them without login. In contrast, Server and Cloud restrict access to authorized users via your organization's network or secure login.
    ->Data Security
    -Public exposes all data openly, risking sensitive info exposure since workbooks are downloadable by anyone. Server/Cloud provide robust security like row level permissions, authentication and encryption to protect proprietary data.

When publishing a Tableau dashboard, what happens to a live connection compared to an extract?
How does this affect data refresh and performance?

    ->Publishing Behavior
    -A live connection retains its direct link to the original database or data source after publishing to Tableau Server/Cloud or Public, querying fresh data each time the dashboard loads. Extracts, however, package a local copy of the data into the workbook or published data source no live link exists post-creation.
    ->Data Refresh
    -Live connections update automatically with every dashboard interaction or view, pulling the latest source data without manual intervention. Extracts require scheduled or manual refreshes on the server updating the snapshot from the source only then.

In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users
can view, interact with, or edit a published dashboard?

    ->Core Feature:The Permissions system is the primary tool accessed via the Actions menu on a workbook, dashboard, or project, opening a dialog to set rules for groups or individual users. Effective permissions show as a grid, inheriting from higher levels but allowing overrides like Deny for stricter control.
    ->Access Capabilities:View: Allows seeing the dashboard .

    Interact: Permits filtering, downloading summary data/images, adding comments.

    Edit/Web Edit: Enables saving changes, publishing updates, or full workbook edits.

If you want users to view a dashboard but not download the underlying data, which permission setting
would you adjust in Tableau Server?

    ->To prevent users from downloading underlying data while allowing dashboard viewing, deny the Download capabilities in Tableau Server permissions. Specifically, target Download Summary Data and Download Full Data for workbooks.
